```python
import opendatasets as od
import pandas
od.download("https://www.kaggle.com/datasets/saadharoon27/hr-analytics-dataset")
```

```python
import pandas as pd
data = pd.read_csv("/content/hr-analytics-dataset/HR_Analytics.csv")
data.head()
```

```python
# question 1
variance = data['MonthlyIncome'].var()
print("Salary Variance =",variance)
stddev = data['MonthlyIncome'].std()
print("Salary Standard deviation =",stddev)
i1 = data['MonthlyIncome'].quantile(0.25)
i2 = data['MonthlyIncome'].quantile(0.75)
interquartile = i2 - i1
print("Salary Interquartile range =",interquartile)
```

```python
# question 2
it_dept = data[data['Department'] == 'IT']
maxexp = it_dept['YearsAtCompany'].max()
minexp = it_dept['YearsAtCompany'].min()
experiencerange = maxexp - minexp
print("The range of experience in the IT department is:",experiencerange)
```

```python
# question 3
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 4))
sns.boxplot(data=data, x='YearsAtCompany')
plt.title('Boxplot')
plt.xlabel('YearsAtCompany')
plt.show()
```

```python
# question 4
numerical_df = data.select_dtypes(include=['float64', 'int64'])
maxstd = numerical_df.std()
highest_std = maxstd.idxmax()
highest_stdvalue = maxstd.max()
print("The feature with the highest standard deviation is:",highest_std)
print("The highest standard deviation value is:",highest_std_value)

```
